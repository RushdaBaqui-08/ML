# Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X,y = make_classification(
    n_samples = 500,
    n_features = 20,
    n_informative = 5,
    n_redundant = 2,
    random_state = 42
)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
svc = SVC()
dtc = DecisionTreeClassifier(max_depth = 3)
```

```python
Voting_Clf = VotingClassifier(
    estimators = [
        ("lr",lr),
        ("svc", svc),
        ("dtc", dtc)
    ]
)
```

```python
Voting_Clf.fit(X_train, y_train)
```

```python
y_pred = Voting_Clf.predict(X_test)

print("accuracy :", accuracy_score(y_pred, y_test))
print("classification_report :",classification_report(y_pred, y_test))
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(
    n_samples = 500,
    n_features = 20,
    n_informative = 5,
    random_state = 42
)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
ling_reg = LinearRegression()
dtr = DecisionTreeRegressor(max_depth = 3)
svr = SVR()
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(
    estimators = [
        ("lr",ling_reg),
        ("dtr", dtr),
        ("svr", svr),
    ]
)
```

```python
vr.fit(X_train, y_train)
```

```python
y_pred = vr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print("r2_score :", r2_score(y_pred, y_test))
```

# Stacking (KFold Method, Blending Method)

```python
# classifier

from sklearn.ensemble import StackingClassifier
```

```python
meta_model = LogisticRegression()

stacking_clf = StackingClassifier(
    estimators = [
        ("lr",lr),
        ("svc", svc),
        ("dtc", dtc)
    ],
    final_estimator = meta_model,
    cv = 5
)
```

```python
stacking_clf.fit(X_train, y_train)
```

```python
y_pred = stacking_clf.predict(X_test)

print("accuracy:", accuracy_score(y_test,y_pred))
```

```python
# regressor

from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import RidgeCV
```

```python
stacking_rg = StackingRegressor(
   estimators = [
        ("lr",ling_reg),
        ("dtr", dtr),
        ("svr", svr),
    ],
    cv = 5
)
```

```python
stacking_rg.fit(X_train, y_train)
```

```python
y_pred_testing = stacking_rg.predict(X_test)
y_pred_training = stacking_rg.predict(X_train)
```

```python
print("r2score_testing:", r2_score(y_test, y_pred_testing))
print("r2score_training:", r2_score(y_train, y_pred_training))
```
